```python
import optuna 

def train_lgb_opt(fold, params, save_model=False):
    df = pd.read_csv("../input/train_folds.csv")
    
    df_train = df[df.kfold != fold].reset_index(drop=True)
    df_valid = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = df_train.drop(columns=["id", "target"], axis=1).values
    ytrain = df_train.target.values
    
    xvalid = df_valid.drop(columns=["id", "target"], axis=1).values
    yvalid = df_valid.target.values
    
    clf = lgb.LGBMClassifier(
        objective="multiclass",
        metric="multi_logloss",
        verbosity=-1,
        boosting_type="gbdt",
        lambda_l1=params["lambda_l1"],
        lambda_l2=params["lambda_l2"],
        num_leaves=params["num_leaves"],
        feature_fraction=params["feature_fraction"],
        bagging_fraction=params["bagging_fraction"],
        bagging_freq=params["bagging_freq"],
        min_child_samples=params["min_child_samples"],
    )
    clf.fit(xtrain, ytrain)
    preds = clf.predict(xvalid)
    accuracy = metrics.accuracy_score(yvalid, preds)
    
    return accuracy

def objective(trial):
    params = {
        "objective": "multiclass",
        "metric": "multi_logloss",
        "verbosity": -1,
        "boosting_type": "gbdt",
        "lambda_l1": trial.suggest_float("lambda_l1", 1e-8, 10.0, log=True),
        "lambda_l2": trial.suggest_float("lambda_l2", 1e-8, 10.0, log=True),
        "num_leaves": trial.suggest_int("num_leaves", 2, 256),
        "feature_fraction": trial.suggest_float("feature_fraction", 0.4, 1.0),
        "bagging_fraction": trial.suggest_float("bagging_fraction", 0.4, 1.0),
        "bagging_freq": trial.suggest_int("bagging_freq", 1, 7),
        "min_child_samples": trial.suggest_int("min_child_samples", 5, 100),
    }
    
    all_accuracy = []
    for f_ in range(5):
        temp_accuracy = train_lgb_opt(f_, params, save_model=False)
        all_accuracy.append(temp_accuracy)
    
    return np.mean(all_accuracy)
```

```python
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=5)

print("Best trial:")
trial_ = study.best_trial

#print(f"Value: {trial_.values}")
print(f"Params: {trial_.params}")

scores = 0
for j in range(5):
    scr = train_lgb_opt(j, trial_.params, save_model=True)
    scores += scr
    
print(scores / 5)
```

```python
import torch
import torch.nn as nn
from sklearn import metrics
import numpy as np
```

```python
class OttoDataset:
    def __init__(self, features, targets):
        self.features = features
        self.targets = targets
        
    def __len__(self):
        return self.features.shape[0]
        
    def __getitem__(self, item):
        return{
            "features": torch.tensor(self.features[item, :], dtype=torch.float),
            "targets": torch.tensor(self.targets[item], dtype=torch.long),
        }
    
```

```python
class Model(nn.Module):
    def __init__(self, num_features, hidden_size, num_targets):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(num_features, hidden_size)
        self.fc2 = nn.Linear(hidden_size, num_targets)
        self.relu = nn.ReLU()
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
    
```

```python
class Engine:
    def __init__(self, model, optimizer, device):
        self.model = model
        self.device = device
        self.optimizer = optimizer

    def train(self, data_loader):
        self.model.train()
        for data in data_loader:
            inputs = data["features"].to(self.device)
            targets = data["targets"].to(self.device)
            
            self.optimizer.zero_grad()
            
            preds = self.model(inputs)
            
            loss = nn.CrossEntropyLoss()(
                preds, 
                targets
            )
            
            loss.backward()
            self.optimizer.step()
            
    def evaluate(self, data_loader):
        final_preds = []
        final_targets = []
        
        self.model.eval()
        with torch.no_grad():
            for data in data_loader:
                inputs = data["features"].to(self.device)
                targets = data["targets"].to(self.device)
                
                preds = self.model(inputs)
                preds = preds.cpu().numpy().tolist()
                targets = data["targets"].cpu().numpy().tolist()
                final_preds.extend(preds)
                final_targets.extend(targets)
            
        return final_preds, final_targets
    
```

```python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
EPOCHS = 10

def run_nn(fold):
    df = pd.read_csv("../input/train_folds.csv")
    
    feature_columns = df.drop(columns=["id", "target", "kfold"], axis=1).columns
    
    df_train = df[df.kfold != fold].reset_index(drop=True)
    df_valid = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = df_train[feature_columns].to_numpy()
    ytrain = df_train["target"].to_numpy()
    
    xvalid = df_valid[feature_columns].to_numpy()
    yvalid = df_valid["target"].to_numpy()
    
    train_dataset = OttoDataset(features=xtrain, targets=ytrain)
    valid_dataset = OttoDataset(features=xvalid, targets=yvalid)
    
    train_loader = torch.utils.data.DataLoader(
        train_dataset, 
        batch_size=104, 
        shuffle=True
    )
    valid_loader = torch.utils.data.DataLoader(
        valid_dataset, 
        batch_size=104
    )
    
    model = Model(
        num_features=xtrain.shape[1], 
        hidden_size=15, 
        num_targets=len(np.unique(ytrain)),
    )
    
    model.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
    eng = Engine(model, optimizer, device=DEVICE)
    
    best_accuracy = 0
    early_stopping_counter = 0
    
    for epoch in range(EPOCHS):
        eng.train(train_loader)
        outputs, targets = eng.evaluate(valid_loader)
        
        outputs = np.array(outputs).argmax(axis=1)
        targets = np.array(targets)
        accuracy = metrics.accuracy_score(targets, outputs)
        
        print(
            f"Fold:{fold}, Epoch:{epoch}, Accuracy={accuracy:.3f}"
        )
        
        if accuracy > best_accuracy:
            best_accuracy = accuracy
        else:
            early_stopping_counter += 1
        
        if early_stopping_counter > 2:
            break
        
        torch.save(model.state_dict(), f"model_fold_{fold}.bin")
            
```

```python
for j in range(5):
    run_nn(j)
```

```python
def predict(fold):
    df = pd.read_csv("../input/test.csv")
    model_path = f"model_fold_{fold}.bin"
    
    feature_columns = df.drop(columns=["id"], axis=1).columns
    
    xtest = df[feature_columns].to_numpy()
    ytest = np.zeros(len(df))
   
    test_dataset = OttoDataset(features=xtest, targets=ytest)
    
    tset_loader = torch.utils.data.DataLoader(
        test_dataset, 
        batch_size=104, 
        shuffle=True
    )
    
    model = Model(
        num_features=xtest.shape[1], 
        hidden_size=15, 
        num_targets=9,
    )
    model.load_state_dict(torch.load(model_path))
    model.to(DEVICE)
    
    eng = Engine(model, device=DEVICE)
    
    outputs, _ = eng.evaluate(test_loader)
    outputs = np.array(outputs)
    print(f"{outputs.shape}")
             
```
